## Number of males and females on board

```python
df['Sex'].value_counts()
```

```python
df['Pclass'].value_counts()
print('Number males from second class on board:', df[(df['Pclass'] == 2) & (df['Sex'] == 'male')].count()['PassengerId'])
```

```python
print("Median:", round(df['Fare'].median(), 2))
print("Std:", round(df['Fare'].std(), 2))
```

```python
print("Passengers survived, age < 30:", df[(df['Age'] < 30) & (df['Survived'] == 1)].count()['PassengerId'])
print("Passengers survived, age > 60:", df[(df['Age'] > 60) & (df['Survived'] == 1)].count()['PassengerId'])
print("Percentage passengers survived, age < 30: ",
      round(df[(df['Age'] < 30) & (df['Survived'] == 1)].count()['PassengerId'] / df[(df['Age'] < 30)].count()['PassengerId'] * 100, 1), "%", sep='')
print("Percentage passengers survived, age > 60: ",
      round(df[(df['Age'] > 60) & (df['Survived'] == 1)].count()['PassengerId'] / df[(df['Age'] > 60)].count()['PassengerId'] * 100, 1), "%", sep='')
```

```python
print("Percentage males survived: ",
      round(df[(df['Sex'] == 'male') & (df['Survived'] == 1)].count()['PassengerId'] /
            df[(df['Sex'] == 'male')].count()['PassengerId'] * 100, 1), "%", sep='')
print("Percentage females survived: ",
      round(df[(df['Sex'] == 'female') & (df['Survived'] == 1)].count()['PassengerId'] /
            df[(df['Sex'] == 'female')].count()['PassengerId'] * 100, 1), "%", sep='')
```

```python
# df['Name'].factorize()
pd.Series(df[df['Sex'] == 'male']['Name'].factorize()[1]).apply(lambda x: x.split()[2]).value_counts()
```
